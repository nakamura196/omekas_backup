```python
#pandasを読み込む
import pandas as pd
```

```python
data_path = "../data"
```

```python
def get_conf_item_set():
    path = data_path + "/item_sets/list_.xlsx"
    input_book = pd.ExcelFile(path)
    #sheet_namesメソッドでExcelブック内の各シートの名前をリストで取得できる
    input_sheet_name = input_book.sheet_names
    #lenでシートの総数を確認
    num_sheet = len(input_sheet_name)
    df_ = input_book.parse(input_sheet_name[0])

    conf = {}

    for index, row in df_.iterrows():
        oid = row["o:id"]
        name = row["o:title"]
        conf[str(oid)] = name

    return conf

conf_item_set = get_conf_item_set()
conf_item_set
```

```python
def get_conf_properties():
    path = data_path + "/properties/list_p.xlsx"
    input_book = pd.ExcelFile(path)
    #sheet_namesメソッドでExcelブック内の各シートの名前をリストで取得できる
    input_sheet_name = input_book.sheet_names
    #lenでシートの総数を確認
    num_sheet = len(input_sheet_name)
    df_ = input_book.parse(input_sheet_name[0])

    conf = {}

    for index, row in df_.iterrows():
        oid = row["o:term"]
        name = row["o:label"]
        conf[str(oid)] = name

    return conf

conf_properties = get_conf_properties()
conf_properties
```

```python
def get_df_item():
    #input file name
    path = data_path + "/items/list.xlsx"
    #xls book Open (xls, xlsxのどちらでも可能)
    input_book = pd.ExcelFile(path)
    #sheet_namesメソッドでExcelブック内の各シートの名前をリストで取得できる
    input_sheet_name = input_book.sheet_names
    #lenでシートの総数を確認
    num_sheet = len(input_sheet_name)
    #シートの数とシートの名前のリストの表示
    # print ("Sheet の数:", num_sheet)
    # print (input_sheet_name)
    df = input_book.parse(input_sheet_name[0])
    return df

df_item = get_df_item()
```

```python
import os
from pandas import json_normalize
from tqdm import tqdm

id_mappings = {}
uuid_set_mappings = {}

data = {}

dc_terms = [
    "dcterms:identifier",
    "dcterms:description",
    "dcterms:type",
    "dcterms:subject",
    "dcterms:extent",
    "dcterms:creator",
    "dcterms:language",
    "dcterms:date",
    "dcterms:publisher",
    "dcterms:rights",
    "dcndl:price",
    "dcndl:volume",
    "archiveshub:note"
]

extra = [
    "foaf:thumbnail"
]

for index, row in tqdm(df_item.iterrows(), total=df_item.shape[0]):

    oid = row["o:id"]
    uuid = row["dcterms:identifier"]
    id_mappings[oid] = uuid

    item_set_labels = []

    item_set = row['o:item_set']
    if pd.isnull(item_set):
        item_sets = ["null"]
    else:
        item_sets = str(item_set).split("|")

    for item_set in item_sets:
        label = conf_item_set[item_set] if item_set in conf_item_set else "null"
        item_set_labels.append(label)

    sort = row["ex:sort"] or row["uterms:sort"]

    item = {
        "system_id": row["dcterms:identifier"],
        "dcterms:title": row["dcterms:title"],
        "public": 1 if row["o:is_public"] else 0,
        "resource_type": "item",
        "parent": "",
        "item set" : "|".join(item_set_labels),
        "sort": sort,
        "item": "",
        "media_source": "",
        "media_rederer": ""
    }

    uuid_set_mappings[uuid] = item_set_labels

    for list in [dc_terms, extra]:

        for term in list:
            if term in row:
                if term == "dcterms:description":
                    if pd.isnull(row[term]):
                        values = []
                    else:
                        values = str(row[term]).split("|")
                    

                    for column_name, item2 in df_item.iteritems():
                        if column_name.startswith("ex:"):
                            column_label = conf_properties[column_name] if column_name in conf_properties else column_name
                            # item[column_name] = item2[index]

                            if not pd.isnull(row[column_name]):
                                values.append(f"{column_label}:{row[column_name]}")

                    item[term] = "|".join(values)
                else:
                    item[term] = row[term]
            else:
                item[term] = ""

    for item_set in item_sets:
        if item_set not in data:
            data[item_set] = []
        data[item_set].append(item)

for item_set_id in tqdm(data):
    # print(conf_item_set[item_set_id])
    label = conf_item_set[item_set_id] if item_set_id in conf_item_set else "null"
    path = "data/" + label + "/item.xlsx"
    os.makedirs(os.path.dirname(path), exist_ok=True)

    # rows = []

    df = json_normalize(data[item_set_id])

    df.to_excel(path, index=False)
```
